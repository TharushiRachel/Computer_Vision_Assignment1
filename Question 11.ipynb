### (a)

By the Convolution Theorem,

$$g(x,y)=f(x,y)*h(x,y) \Rightarrow G(u,v)=F(u,v)H(u,v).$$

So, convolution in the spatial domain becomes multiplication in the frequency domain.

### (b)

**Averaging (box) filter:** low-pass filter. It smooths the image and reduces high frequencies.

**Gaussian filter:** smooth low-pass filter. It blurs the image gently and suppresses high frequencies smoothly.

**Laplacian filter:** high-pass filter. It enhances edges and fine details by emphasizing high frequencies.

### (c)

Gaussian filtering avoids ringing because its frequency response changes smoothly. An ideal low-pass filter has a sharp cutoff, which causes oscillations and ringing near edges.

### (d)

The **Gaussian filter** is best for high-frequency noise reduction.

In the spatial domain, it smooths the image gently.

In the frequency domain, it suppresses high frequencies smoothly.

The box filter also reduces noise, but Gaussian gives fewer artifacts. The Laplacian is not suitable because it increases high-frequency noise.
