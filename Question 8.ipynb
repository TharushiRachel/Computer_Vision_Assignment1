# Question 8

`salt&paper.png` is corrupted with salt and pepper noise.

(a) Apply Gaussian smoothing.

(b) Apply median filtering.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/salt&paper.png', cv2.IMREAD_GRAYSCALE)

gaussian = cv2.GaussianBlur(img, (5, 5), 0)
median = cv2.medianBlur(img, 5)

fig, ax = plt.subplots(1, 3, figsize=(12, 4))
for a, im, title in zip(ax, [img, gaussian, median], ['Original', 'Gaussian smoothing', 'Median filtering']):
    a.imshow(im, cmap='gray')
    a.set_title(title)
    a.axis('off')
plt.tight_layout()
plt.show()
```
